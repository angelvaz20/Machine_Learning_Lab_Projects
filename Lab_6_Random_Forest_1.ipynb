```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
df=pd.read_csv('student_productivity_distraction_dataset_20000.csv')
df
```

```python
print(df.shape)
```

```python
# eliminate the null data
df = df.dropna()
print(df.shape)
```

```python
#split dataset in features and target variable
x_data=df[['study_hours_per_day','sleep_hours','phone_usage_hours',
           'gaming_hours','breaks_per_day','coffee_intake_mg',
           'exercise_minutes','attendance_percentage','stress_level']]
y_data = pd.cut(df['final_grade'],
                bins=[0, 60, 100],
                labels=['Fail','Pass'])
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3, random_state=1)
```

```python
# Train a single Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train,y_train)

# Calculate the accuracy and assign it to dt_accuracy
# Train a single Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# Calculate the accuracy and assign it to dt_accuracy
dt_accuracy = dt_model.score(X_test, y_test)

print("Decision Tree Test Accuracy:", f"{dt_accuracy * 100:.2f}%")
```

```python
rf_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    random_state=42
)

rf_model.fit(X_train, y_train)
rf_accuracy = rf_model.score(X_test, y_test)

print("Random Forest Test Accuracy:", f"{rf_accuracy * 100:.2f}%")
```

```python
rf_model = RandomForestClassifier(n_estimators=200, max_features=10, random_state=42)
rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)

precision_rf = metrics.precision_score(y_test, y_pred_rf, pos_label='Pass')
recall_rf    = metrics.recall_score(y_test, y_pred_rf, pos_label='Pass')
f1_rf        = metrics.f1_score(y_test, y_pred_rf, pos_label='Pass')

print("RF Precision:", precision_rf)
print("RF Recall:", recall_rf)
print("RF F1-Score:", f1_rf)
```

# Summary
This project analyzes data from 20,000 students to examine how daily habits and behaviors relate to academic success. It looks at factors such as study time, sleep, phone usage, gaming, breaks, caffeine intake, exercise, attendance, and stress levels to predict whether a student is likely to pass or fail. Instead of predicting exact grades, the program simplifies outcomes into two pass or fail categories and uses machine learning models to identify patterns in the data. After training on past student information, the system tests its accuracy and selects the stronger model to make predictions for new students based on their habits. It also determines which factors have the greatest influence on academic performance, helping highlight behaviors that may contribute most to student success.
