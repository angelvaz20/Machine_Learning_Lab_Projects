```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("tweets.csv")
# Keep only the columns we need
df = df[["text", "target"]]

# Remove missing values just in case
df = df.dropna()

# Get the tweets and labels
tweets = df["text"]
labels = df["target"]
df
```

```python
print("=== STEP 2: Dataset loaded ===")
print(f"Number of tweets: {len(tweets)}")
print()
```

```python
vectorizer = CountVectorizer()
```

```python
X = vectorizer.fit_transform(tweets)
```

```python
print("=== STEP 3: Words converted to numbers ===")
print(f"Number of unique words found: {len(vectorizer.get_feature_names_out())}")
print()
```

```python
model = MultinomialNB()
model.fit(X, labels)
```

```python
from sklearn.model_selection import train_test_split
# Split the data (this is IMPORTANT for reliability)

X_train, X_test, y_train, y_test = train_test_split(
    tweets, labels, test_size=0.2, random_state=42
)
```

```python
# Vectorize AGAIN (fit only on training data)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
# Train model on training data
model.fit(X_train_vec, y_train)

# Make predictions on TEST data
y_pred = model.predict(X_test_vec)
```

```python
# Results
# 1. Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
print()

# 2. Classification Report
print("Classification Report:")
print(classification_report(y_test, y_pred))
print()
```

```python
# Testing Model on new tweets not seen before
new_tweets = [
    "Massive fire destroys several homes in California",
    "I love this new song so much",
    "Earthquales shakes downtown area and damages buldings",
    "Had a great lunch with my friends today",
    "Flood warning issued after heavy rain in the city ",
    "I loved reading those books from the book club"

]
# Convert new tweets into numbers using the SAME vectorizer
X_new = vectorizer.transform(new_tweets)

# Ask the model to make predictions
predictions = model.predict(X_new)

# Get probability scores
probabilities = model.predict_proba(X_new)
```

```python
print("=== 3Predictions on new tweets ===")
print()

for i, tweet in enumerate(new_tweets):
    # predictions[i] is either 1 (disaster) or 0 (not disaster)
    result = "DISASTER" if predictions[i] == 1 else "NOT DISASTER"

    # probabilities[i][0] = probability of NOT disaster
    # probabilities[i][1] = probability of disaster
    prob_disaster = probabilities[i][1] * 100
    prob_notdisaster = probabilities[i][0] * 100

    print(f"Tweet: '{tweet}'")
    print(f"  --> Prediction      : {result}")
    print(f"  --> Disaster chance : {prob_disaster:.1f}%")
    print(f"  --> Not disaster    : {prob_notdisaster:.1f}%")
    print()
```
